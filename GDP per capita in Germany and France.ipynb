# GDP per capita in Germany and France (2000 - 2020)

### constant 2010 USD

### source: The Word Bank

importing packages


```python
from datetime import datetime
from matplotlib import pyplot as plt
from IPython.display import display
import pandas as pd
import pandas_datareader.wb as wb
```

getting GDP per capita data from The Word Bank

```python
start = datetime(2000, 1, 1)
end = datetime(2020, 1, 1)

gdp_data_de = wb.download(indicator='NY.GDP.PCAP.KD', country=['DE'], start=start, end=end)
gdp_data_de.rename(columns={'NY.GDP.PCAP.KD': 'GDP_per_capita'}, inplace=True)
display(gdp_data_de)

gdp_data_fr = wb.download(indicator='NY.GDP.PCAP.KD', country=['FR'], start=start, end=end)
gdp_data_fr.rename(columns={'NY.GDP.PCAP.KD': 'GDP_per_capita'}, inplace=True)
display(gdp_data_fr)
```

plotting the graph

```python
reversed_gdp_data_de = gdp_data_de.iloc[::-1]
reversed_gdp_data_fr = gdp_data_fr.iloc[::-1]

plt.figure(figsize=(15,10))
ax = plt.subplot()
plt.plot(range(len(reversed_gdp_data_de)), reversed_gdp_data_de['GDP_per_capita'], linestyle='--', marker='o', color='red', label='Germany')
plt.plot(range(len(reversed_gdp_data_fr)), reversed_gdp_data_fr['GDP_per_capita'], linestyle='--', marker='o', color='blue', label='France')
ax.set_xticks(range(len(reversed_gdp_data_de)))
ax.set_xticklabels(list(range(2000, 2021)))
plt.xlabel('year')
plt.ylabel('GDP per capita (constant 2010 $)')
plt.title('GDP per capita in Germany and France')
plt.legend()
plt.show()
```
